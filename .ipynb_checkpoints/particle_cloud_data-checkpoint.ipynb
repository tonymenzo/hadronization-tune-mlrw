```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Insert PATH to datasets
PATH_sim = '<PATH>/sim_hadrons_a_1.5_b_0.6_nevents_1e6.npy'
PATH_exp = '<PATH>/exp_hadrons_a_0.6_b_1.5_nevents_1e6.npy'

events_sim = np.load(PATH_sim)
events_exp = np.load(PATH_exp)
```

```python
print('Simulated dataset shape:', events_sim.shape)
print('Experimental dataset shape:', events_exp.shape)
```

```python
# First index is at the event level (one full hadronization event). The event is zero padded to a fixed length of 100
print(events_sim[0])
```

```python
# The second index is at the hadron level (fragmented hadron four-momenta - the convention is [p_x, p_y, p_z, E])
print(events_sim[0,0])
```

```python
# Plot the first emission p_z distribution
fig, ax = plt.subplots(1,1,figsize = (5,4))
_, bins, _ = ax.hist(events_sim[:,0,0], 20, label = r'$\mathrm{Sim.}$', alpha = 0.6)
ax.hist(events_exp[:,0,0], bins, label = r'$\mathrm{Exp.}$', alpha = 0.6)
ax.set_xlabel(r'$p_z$')
ax.legend(frameon = False)
fig.tight_layout()
```

```python
# Plot the first emission p_T distribution
fig, ax = plt.subplots(1,1,figsize = (5,4))
_, bins, _ = ax.hist(np.sqrt(events_sim[:,0,1]**2 + events_sim[:,0,2]**2), 20, label = r'$\mathrm{Sim.}$', alpha = 0.6)
ax.hist(np.sqrt(events_exp[:,0,1]**2 + events_exp[:,0,2]**2), bins, label = r'$\mathrm{Exp.}$', alpha = 0.6)
ax.set_xlabel(r'$p_T$')
ax.legend(frameon = False)
fig.tight_layout()
```

```python
# Plot the first emission E distribution
fig, ax = plt.subplots(1,1,figsize = (5,4))
_, bins, _ = ax.hist(events_sim[:,0,3], 20, label = r'$\mathrm{Sim.}$', alpha = 0.6)
ax.hist(events_exp[:,0,3], bins, label = r'$\mathrm{Exp.}$', alpha = 0.6)
ax.set_xlabel(r'$E$')
ax.legend(frameon = False)
fig.tight_layout()
```

```python
# Compute the event multiplicities
mult_sim = np.zeros(events_sim.shape[0])
mult_exp = np.zeros(events_exp.shape[0])
for i in range(events_sim.shape[0]):
    # Use the energy (which is always positive) to extract multiplicity
    mult_sim[i] += len(events_sim[i,:,3][events_sim[i,:,3] > 0.0])
    mult_exp[i] += len(events_exp[i,:,3][events_exp[i,:,3] > 0.0])
```

```python
# Plot the first emission E distribution
fig, ax = plt.subplots(1,1,figsize = (5,4))
_, bins, _ = ax.hist(mult_sim, 20, label = r'$\mathrm{Sim.}$', alpha = 0.6)
ax.hist(mult_exp, bins, label = r'$\mathrm{Exp.}$', alpha = 0.6)
ax.set_xlabel(r'$N_h$')
ax.legend(frameon = False)
fig.tight_layout()
```
